```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
dfgoods=[]
dfname=[]
dffac=[]
dfcateg=[]
dfprice=[]
dfimage=[]
```

```python
for page in range(0,7):
    response = requests.get("https://online.carrefour.com.tw/zh/search?q=%E8%A1%9B%E7%94%9F%E7%B4%99&start=" + str(page*24))
    soup = BeautifulSoup(response.text, "html.parser")
    elements = soup.find_all(class_='gtm-product-alink')
    for element in elements:
        dfgoods.append('https://online.carrefour.com.tw'+element['href'])
        print('https://online.carrefour.com.tw'+element['href'])

print(dfgoods)
```

```python
dict = {'網址':dfgoods}
print(dict)
df = pd.DataFrame(dict) 
df.to_csv('carrefour_link.csv')
```
