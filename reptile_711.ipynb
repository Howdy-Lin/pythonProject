```python
# 載入需要的套件
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
dfname=[]
dffac=[]
dfcateg=[]
dfprice=[]
dfimage=[]
i=1
# 開啟瀏覽器視窗(Chrome)
# 方法一：執行前需開啟chromedriver.exe且與執行檔在同一個工作目錄
driver = webdriver.Chrome()
driver.get("https://shop.7-11.com.tw/shop/rui001.faces") # 更改網址以前往不同網頁
# 定位搜尋框
element = driver.find_element_by_class_name('ac_input')
# 傳入字串
element.send_keys('衛生紙')
element.send_keys(Keys.ENTER)
url = BeautifulSoup(driver.page_source, 'html.parser')   #抓下當前使用網址
catalog = url.find(class_='pagelist1')
next_page = catalog.find_all('a')   #抓含有下一頁超連結的element

elements = url.find_all(class_='prodM')

print(next_page)
next_page=str(next_page)
while  ('下一頁'in next_page)==True:

    for element in elements:       
        dfgoods.append('https://shop.7-11.com.tw/shop/'+element['href'])
        print('https://shop.7-11.com.tw/shop/'+element['href'])        
        element1 = element.find(class_='wordwidth').getText().strip()
        pics = element.find_all('img')
        image_links = [pic.get("src") for pic in pics]
        image_links ='https://shop.7-11.com.tw'+image_links[0]   
        dfimage.append(image_links)
        url1=image_links
        r=requests.get(url1)
        with open(f'./images/{i}'+'.jpg','wb') as f:
            f.write(r.content)
            i=i+1
        price = element.find("b").getText().strip()
        dfname.append(element1)
        dfprice.append(price)
        if('】'in element1)==True:
            element1=element1.split('】')
            name=element1[1]
            fac=element1[0]
            fac=''.join(fac)
            fac=fac.split('【')
            fac=fac[1]        
            dffac.append(fac)
            if ('濕式' in name)==True:
                categ = name.partition('濕式')
                categ = list(categ)
                print(categ[1])
                dfcateg.append(categ[1])
            elif ('抽取' in name)==True:
                categ = name.partition('抽取')
                categ = list(categ)
                print(categ[1])
                dfcateg.append(categ[1])
            elif ('平版' in name)==True:
                categ = name.partition('平版')
                categ = list(categ)
                print(categ[1])
                dfcateg.append(categ[1])
            elif ('捲筒' in name)==True:
                categ = name.partition('捲筒')
                categ = list(categ)
                print(categ[1])
                dfcateg.append(categ[1])
            else:
                dfcateg.append('')
        else:
            fac=''
            dffac.append('')
            dfcateg.append('')

   # print(title1,title2)
    catalog = url.find(class_='pagelist1')
    next_page = catalog.find_all('a',string='下一頁')   #抓含有下一頁超連結的element
    print(next_page)
    for href in next_page:            #要用for迴圈才可取得next_page裡的element
        url = href.get('href')
        driver.get(url)
        url = BeautifulSoup(driver.page_source, 'html.parser')   #抓下當前使用網址的html
        elements = url.find_all(class_='prodM')  
    url = BeautifulSoup(driver.current_url) 
    url=url.find('p')
    url=url.getText()
    print(url)
    url = BeautifulSoup(driver.page_source, 'html.parser')   #抓下當前使用網址的html
    catalog = url.find(class_='pagelist1')
    next_page = catalog.find_all('a',string='下一頁')   #抓含有下一頁超連結的element
    next_page = str(next_page)   #建立新的next_page 讓while迴圈可以繼續跑
```
