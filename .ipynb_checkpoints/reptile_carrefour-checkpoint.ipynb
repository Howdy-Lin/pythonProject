```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
dfname=[]
dffac=[]
dfcateg=[]
dfprice=[]
dfimage=[]
```

```python
for page in range(0,7):
    response = requests.get("https://online.carrefour.com.tw/zh/search?q=%E8%A1%9B%E7%94%9F%E7%B4%99&start=" + str(page*24))
    soup = BeautifulSoup(response.text, "html.parser")
    elements = soup.find_all(class_='commodity-desc')
    for element in elements:
        name=element.find('a').getText().strip()  
        dfname.append(name)
        print(name)
        if ('家樂福' in name)==True:
            categ = name.partition('家樂福')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('五月花' in name)==True:
            categ = name.partition('五月花')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])       
        elif ('情人' in name)==True:
            categ = name.partition('情人')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1]) 
        elif ('唯潔雅' in name)==True:
            categ = name.partition('唯潔雅')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1]) 
        elif ('舒潔' in name)==True:
            categ = name.partition('舒潔')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1]) 
        elif ('百吉' in name)==True:
            categ = name.partition('百吉')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('雪柔' in name)==True:
            categ = name.partition('雪柔')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('柔情' in name)==True:
            categ = name.partition('柔情')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])          
        elif ('舒適' in name)==True:
            categ = name.partition('舒適')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])  
        elif ('優活' in name)==True:
            categ = name.partition('優活')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('金舒適' in name)==True:
            categ = name.partition('金舒適')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('春風' in name)==True:
            categ = name.partition('春風')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('得意' in name)==True:
            categ = name.partition('得意')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('可麗舒' in name)==True:
            categ = name.partition('可麗舒')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('蒲公英' in name)==True:
            categ = name.partition('蒲公英')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])
        elif ('倍潔雅' in name)==True:
            categ = name.partition('倍潔雅')
            categ = list(categ)
            print(categ[1])
            dffac.append(categ[1])            
        else :
            print('')
            dffac.append('')
#print(dfname)
```

```python

for page in range(0,7):
    response = requests.get("https://online.carrefour.com.tw/zh/search?q=%E8%A1%9B%E7%94%9F%E7%B4%99&start=" + str(page*24))
    soup = BeautifulSoup(response.text, "html.parser")
    elements = soup.find_all(class_='commodity-desc')
    for element in elements:
        name=element.find('a').getText().strip()  
        #print(name)
        if ('抽取' in name)==True:
            #print(name)
            categ1 = name.partition('抽取')
            #print(categ1)
            categ1 = list(categ1)
            dfcateg.append(categ1[1])
        elif('平版' in name)==True:
            categ1 = name.partition('平版')
            #print(categ1)
            dfcateg.append(categ1[1])
        elif('濕式' in name)==True:
            categ1 = name.partition('濕式')
            #print(categ1)                        
            dfcateg.append(categ1[1])  
        elif('捲筒' in name)==True:
            categ1 = name.partition('捲筒')
            #print(categ1)                        
            dfcateg.append(categ1[1])    
        elif('紙巾' in name)==True:
            categ1 = name.partition('紙巾')
            #print(categ1)                        
            dfcateg.append(categ1[1])                
        else:
            print(name)
            dfcateg.append('')
print(dfcateg)
```

```python

for page in range(0,7):
    response = requests.get("https://online.carrefour.com.tw/zh/search?q=%E8%A1%9B%E7%94%9F%E7%B4%99&start=" + str(page*24))
    soup = BeautifulSoup(response.text, "html.parser")
    elements = soup.find_all(class_='current-price')
    for element in elements:
        price=element.find('em').getText().strip()
        dfprice.append(price)
print(dfprice)
```
